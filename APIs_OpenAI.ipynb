```python
from google.colab import userdata
from openai import OpenAI

api_key = userdata.get("OPENAI_API_KEY")
```

#API do CHAT GPT para gerar resposta de perguntas

```python
client = OpenAI(api_key=api_key) # Chave de API
# Modelo de pergunta
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Você é um professor de progamação especializado em inteligencia artificial"},
        {"role": "system", "content": "Explique o conceito de inteligência artificial de forma simples para desenvolvedores."}
        ],
        max_tokens=150,
        temperature=0.5
    )

print(response)
```

```python
response.choices[0].message.content
```

#API do DALL-E para imagens



```python
response = client.images.generate(
    model = "dall-e-2", # Certifique-se de usar o modelo correto para DALL-E
    prompt = "uma ilustação de um gato astronauta no espaço, estilo cartoon",
    n=5, # Numero de imagens geradas
    size="1024x1024" # Tamanho da imagem
)

# Gerar ULR da imagem gerada
print(response.data[0].url)
```

#API do GPT para geração de código


```python
#Gerando código a partir de um prompt
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Você é um desenvolvedor Python experiente."},
        {"role": "user", "content": "Escreva uma função em Python que aceite um número inteiro e retorne True se número for primo, caso contrário, retorne False"}
    ],
    max_tokens=500,
    temperature=0.5
)

print(response.choices[0].message.content.strip())
```

# API do Whisper transcrever áudio


```python
# Caminho para o arquivo de áudio que você deseja transcrever
audio_file_patg = '/content/audio.wav'

audio_file= open(audio_file_patg, "rb")
transcription = client.audio.transcriptions.create(
    model="whisper-1",
    file=audio_file
)
print(transcription.text)
```

```python
# Tradução

audio_file= open(audio_file_patg, "rb")
translation = client.audio.translations.create(
    model="whisper-1",
    file=audio_file
)
print(translation.text)
```

# API do Embeddings converter textos em vetores numéricos

```python
# Texto para os quais você deseja gera embeddings
texts = ["Aprendizado de máquina é o estudo de algoritimos de computador que melhoram autimaticamente com a experiência",
        "inteligencia artificila é a ciencias de fazer as maquinas fazerem coisas que exigiriam inteligencia se feita por humanos"]

# Gerando embeddings para cada texto
embeddings = []
for text in texts:
  response = client.embeddings.create(
      input=text,
      model="text-embedding-3-small" # Substitua pelo modelo de embeddings disponivel
  )
  embeddings.append(print(response.data[0].embedding))

# Agora você tem uma lista de vetores de em beddings para cada texto
embeddings
```
